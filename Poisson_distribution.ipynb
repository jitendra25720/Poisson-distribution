```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as st
import matplotlib.animation as animation
```

```python
k=100

def update(curr):
    if curr == k:
        a.event_source.stop()
    plt.cla()
    x=np.linspace(1,100,100)
    pois = st.poisson.pmf(x,curr)
    plt.plot(x, pois,'-o', c='green')
    plt.gca().set_title('Poisson distribution, Sample size=100, mean $\mu$={}'.format(curr))
    plt.gca().fill_between(x,min(pois), pois, facecolor = 'green', alpha=0.15)
    plt.gca().set_ylabel('Probability')
    plt.gca().set_xlabel('Value')
    plt.plot()
    if curr==51:
        plt.pause(3)
```

```python
fig = plt.figure()
a = animation.FuncAnimation(fig, update, frames=1000, interval=150)
```
